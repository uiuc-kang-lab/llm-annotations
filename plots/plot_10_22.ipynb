# Global warming dataset

```python
import pandas as pd

llm_only = 0.23169225607361227
llm_human = pd.read_csv("global_warming_llm_human_confidence.csv", 
                        names=["dataset", "sampling_rate", "error", 
                               "llm_cost", "human_cost"])
human_sampling = pd.read_csv("global_warming_smapling.csv",
                             names=["dataset", "sampling_rate", "error",
                                    "llm_cost", "human_cost"])
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(3,2))
plt.plot(llm_human["human_cost"], llm_human["error"]*100, label="LLM+Human")
plt.axhline(y=llm_only*100, color='r', linestyle='--', label="LLM only")
plt.legend(frameon=False, bbox_to_anchor=(1, 1))
plt.xlabel("# Human Labels")
plt.ylabel("Relative Error (%)")
plt.savefig("global_warming_llm_human.pdf", bbox_inches='tight')
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(3,2))
plt.plot(human_sampling["human_cost"][10:], human_sampling["error"][10:]*100, label="Uniform sampling")
plt.axhline(y=llm_only*100, color='r', linestyle='--', label="LLM")
plt.legend(frameon=False, bbox_to_anchor=(1, 1))
plt.xlabel("# Human Labels")
plt.ylabel("Relative Error (%)")
plt.savefig("global_warming_sampling.pdf", bbox_inches='tight')
```
