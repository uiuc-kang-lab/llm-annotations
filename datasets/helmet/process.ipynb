```python
import pandas as pd
raw_data = pd.read_csv("../raw/helmet/all.csv")
raw_data["gold_label"] = 0
raw_data.loc[raw_data["helmet_positive"] == 1, "gold_label"] = 1
raw_data.loc[raw_data["helmet_negative"] == 1, "gold_label"] = 2
```

```python
gpt_data = pd.read_csv("../raw/helmet/gpt_intermediate.csv")
```

```python
merged_data = pd.merge(raw_data, gpt_data, on="cpsc_case_number")
```

```python
to_label = {
    "helmet not mentioned": 0,
    "wearing helmet": 1,
    "not wearing helmet": 2
}

merged_data["gpt_label"] = merged_data["helmet_status"].map(to_label)
```

```python
import numpy as np
np.sum(merged_data["gold_label"] == merged_data["gpt_label"]) / len(merged_data)
```

```python
merged_data.head(1)
```

```python
data = pd.DataFrame({
    "cpsc_case_number": merged_data["cpsc_case_number"],
    "narrative": merged_data["narrative_1"],
    "gold_label": merged_data["gold_label"],
    "mode": merged_data["mode"],
    "gpt_label": merged_data["gpt_label"]
})

data.to_csv("data.csv", index=False)

data
```
