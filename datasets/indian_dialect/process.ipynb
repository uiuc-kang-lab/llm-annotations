```python
import pandas as pd
gpt_labels = pd.read_csv("../raw/indian_dialect/answer-chatgpt", sep="\t", header=None)
human_labels = pd.read_csv("../raw/indian_dialect/indian_english_dialect.csv")
gpt_labels
```

```python
human_labels
```

```python
to_label = {}
for i, label in enumerate(human_labels["Feature"].unique()):
    to_label[label] = i
to_label
```

```python
import numpy as np
accuracy = np.sum(gpt_labels[1] == human_labels["Feature"]) / len(gpt_labels)
accuracy
```

```python
gpt_labels["label"] = gpt_labels[1].map(to_label)
human_labels["label"] = human_labels["Feature"].map(to_label)
data = pd.DataFrame({
    "sentence": human_labels["Example"],
    "gold_label": human_labels["label"],
    "gpt_label": gpt_labels["label"],
    "gold_label_text": human_labels["Feature"]
})

data.to_csv("data.csv")

data
```
